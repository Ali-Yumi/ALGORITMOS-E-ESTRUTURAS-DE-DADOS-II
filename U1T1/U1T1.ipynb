```python
#Exercise 2.3
# Define Caesar Cipher change
def caesar_cipher_shift(text, shift):
    result = []
    
    for char in text:
        if char.isalpha():  # Only change the alphabet characteres 
            base = ord('A')
            shifted_char = chr((ord(char) - base + shift) % 26 + base)
            result.append(shifted_char)
        else:
            result.append(char)  # Non-alphabetic characters stay unchanged
    
    return ''.join(result)
originalText = "OCZ XJMMZXO VINRZM"

# Apply the Caesar change (change by 5)
shift_value = 5
final_text = originalText

final_text = caesar_cipher_shift(final_text, shift_value)

final_text
```

```python
#Exercise 2.4 plot
import matplotlib.pyplot as plt

# Outcome probabilities
outcomes = [1, 2, 3, 4, 5, 6, 7, 8]
probabilities = [0.1, 0.15, 0.2, 0.21, 0.17, 0.09, 0.06, 0.02]

# Probability mass function plot
plt.figure(figsize=(8, 6))
plt.bar(outcomes, probabilities, color='pink')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Probability Mass')
plt.grid(axis='y')
plt.tight_layout()

# Cumulative distribution function plot
cumulative_probabilities = [sum(probabilities[:i+1]) for i in range(len(probabilities))]
plt.figure(figsize=(8, 6))
plt.plot(outcomes, cumulative_probabilities, marker='o', linestyle='-', color='purple')
plt.xlabel('Outcome')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution')
plt.grid(axis='y')
plt.tight_layout()

plt.show()
```

```python
#Exercise 4.1
import networkx as nx
import matplotlib.pyplot as plt
G = nx.read_edgelist("Q5A5.txt")
nodes = nx.algorithms.bipartite.basic.sets(G)
# to identify the bipartite graph
node_neighbors = {n: len(set(G.neighbors(n))) for n in nodes[0]}
#"for n in nodes[0]" to analasy every node in the first group of the bipartite graph
#"len(set(G.neighbors(n)))" to calculate every unique neighbors
maxnode = max(node_neighbors, key = node_neighbors.get)
#"max" to find the highest element of the sequel, but, we need to use the "key" to especify the function that he need to use
print(maxnode, node_neighbors[maxnode])
# second part
node_neighbors = {n: len(set(G.neighbors(n))) for n in nodes[1]}
maxnode = max(node_neighbors, key = node_neighbors.get)
print(maxnode, node_neighbors[maxnode])
#just to see better what happend
pos = nx.spring_layout(G,k=0.2)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=300, width=0.7)
plt.show()
```

```python
#Exercise 4.2
import networkx as nx
from collections import defaultdict
# Load the data. We need to import as multigraph, or networkx will collapse the edges.
# We also need to make sure to import the edge type information.
G = nx.read_edgelist("Q6A6.txt", create_using = nx.MultiGraph(), data = [("edgetype", str),])
# Determine how many layers we have
layers = len(set(e[2]["edgetype"] for e in G.edges(data = True))) - 1
# If this is a chain coupling, all nodes can have at most two coupling edges
node_cedges = defaultdict(int)
for e in G.edges(data = True):
   if e[2]["edgetype"] == "C":
      node_cedges[e[0]] += 1
      node_cedges[e[1]] += 1

if max(node_cedges.values()) == 2:
   print("Chain coupling")
# If this is a clique coupling, all nodes must have more than one coupling edge
if min(node_cedges.values()) > 1:
   print("Clique coupling")
# If this is a star coupling, all nodes must have either one edge, or |L| - 1 coupling edges
if set([1, layers - 1]) == set(node_cedges.values()):
   print("Star coupling")
```
